```python
import pandas as pd
import matplotlib.pyplot as plt

employee_data = pd.read_csv('/content/employees dataset.csv')

# first few rows to understand the data structure
print(employee_data.head())

# Age distribution
plt.hist(employee_data['age'], bins=20, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()

# KPI Met > 80% distribution
plt.bar(['KPI Met > 80%', 'KPI Not Met > 80%'], employee_data['KPIs_met_more_than_80'].value_counts(), color=['green', 'red'])
plt.xlabel('KPI Met More Than 80%')
plt.ylabel('Count')
plt.title('KPI Met More Than 80% Distribution')
plt.show()

# Relationship between previous year rating and awards won
plt.scatter(employee_data['previous_year_rating'], employee_data['awards_won'], color='orange')
plt.xlabel('Previous Year Rating')
plt.ylabel('Awards Won')
plt.title('Relationship between Previous Year Rating and Awards Won')
plt.show()


```
