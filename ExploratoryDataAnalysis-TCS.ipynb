```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# data is in a CSV file named '/content/TCS Dataset.csv'
data = pd.read_csv('/content/TCS Dataset.csv')

print(data.head())

# Basic information about the dataset
print(data.info())

#statistics of the numerical columns
print(data.describe())

# missing values
print(data.isnull().sum())

# Visualizing the trends and patterns in the data
plt.figure(figsize=(12, 6))

#'Close' prices over time
plt.plot(data['Date'], data['Close'])
plt.title('TCS Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks([])

plt.show()

# Pairplot for an overall view of relationships between numerical columns
sns.pairplot(data)
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
